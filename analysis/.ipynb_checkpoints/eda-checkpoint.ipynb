# Import the Data file and Added GDP Trend Feature

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gdp_data.csv")

df["gdp_growth"]= df.groupby("country")["gdp"].pct_change()*100

```

## Insights for TOP 10 Economics

```python
df_latest = df[df["year"]==df["year"].max()]

top10 = df_latest.sort_values("gdp",ascending=False).head(10)
```
